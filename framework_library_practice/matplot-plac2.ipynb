```python
import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(0, 10, 100)
y= np.sin(x)
z= np.cos(x)

#plotting the data
#sin()
"""
plt.plot(x,y)
plt.show()
"""
#cos()
plt.plot(x,z)
plt.xlabel("angle")
plt.ylabel("cos(x)")
plt.title("cos wave")
plt.show()
```

```python
x= np.linspace(-10, 10, 20)
y= x**2
plt.plot(x,y, 'g.')
plt.show()
```

```python
x= np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

bar graph

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
languages= ['englisg', 'french', 'latin', 'german', 'spanish']
people= [100, 50, 150, 40, 70]

ax.bar(languages, people)
plt.xlabel("languages")
plt.ylabel('num of ppl')

plt.show()
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
languages= ['englisg', 'french', 'latin', 'german', 'spanish']
people= [100, 50, 150, 40, 70]

ax.bar(people, labels= languages, autopct='%1.1f%%')


plt.show()
```

3d scatter plot


```python
fig3= plt.figure()
ax= plt.axes(projection= '3d')

z= 20*np.random.random(100)
x= np.sin(z)
y= np.cos(z)

ax.scatter(x, y, z, c=z, cmap= 'Blues')
plt.show()
```
